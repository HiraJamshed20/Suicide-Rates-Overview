## Hypothesis testing - Gender

```python
genderdf = pd.DataFrame(sdf, columns = ['country', 'year','sex', 'suicides_no', 'population'])
genderdf = genderdf[genderdf['year'] == 2000]
bool_arr = ((genderdf['country'] == 'Japan'))
genderdf = genderdf[bool_arr]
genderdf.groupby('sex').sum().head()
```

```python
############################################ HYPOTHESIS TESTING FOR GENDER ############################################
tot_pop = genderdf['population'].sum()
tot_dead = genderdf['suicides_no'].sum()
m_dead = genderdf['suicides_no'][1]
f_dead = genderdf['suicides_no'][0]
#print(msuicide, fsuicide)
curr_pop = 0;
curr_pop_alive = 0;
curr_pop_dead = 0;
curr_f_dead = 0;
curr_m_dead = 0;

picked_f = []
picked_m = []
picked_alive = []
```

```python
########################################### RESETTING GENDER COUNTERS #########################################
def reset_sex():
    
    global curr_pop_alive
    global curr_pop_dead
    global curr_f_dead
    global curr_m_dead
    global curr_pop
    global tot_pop
    global tot_dead
    global f_dead
    global m_dead
    
    global picked_f
    global picked_m
    global picked_alive
    
    curr_pop = tot_pop
    curr_pop_alive = tot_pop - tot_dead
    curr_pop_dead = tot_dead
    curr_f_dead = f_dead
    curr_m_dead = m_dead
    
    picked_m = []
    picked_f = []
    picked_alive = []

#################################################################################################################

def draw_sexy():
    global curr_pop_alive
    global curr_pop_dead
    global curr_f_dead
    global curr_m_dead
    global curr_pop
    
    global picked_f
    global picked_m
    global picked_alive
    picked = 0
    
    #first of all pick a random number between 0 to 100.
    randy = random()
    randy = 0 + (randy * (100 - 0))
    # Now we check if this individual is going to commit suicide or nah
    # Chance for committing suicide is  = curr_people who died/ curr_popsize
    suicide_thresh = (curr_pop_dead/curr_pop) *100
    #print("thresh_s :",suicide_thresh, "randy:", randy)
    if randy <= suicide_thresh: 
        #we picked a dead person Prolly (boi)
        picked =1
        #we need to determine if this is a dead pee pee or a vee vee
        peepee_thresh = (curr_m_dead/ (curr_pop_dead)) * 100
        
        randy_2 = random()
        randy_2 = 0 + (randy_2 * (100 - 0))
        
        if randy_2 <= peepee_thresh:
            #this is a male, decrement male dead pop and total pop
            curr_m_dead = curr_m_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_m.append(picked)
            
        else:
            #this is a female, decrement female dead pop and total pop
            curr_f_dead = curr_f_dead - 1
            curr_pop = curr_pop - 1
            curr_pop_dead = curr_pop_dead -1
            picked_f.append(picked)
    else:
        # we picked an alive person wuhuuuuu no one dies
        # update alive counter and total pop counter
        curr_pop_alive = curr_pop_alive-1
        curr_pop = curr_pop - 1
        picked_alive.append(picked)
        
            
################### testing ##########
# reset_sex()
# print("curr_pop:",curr_pop,"\nc_f_dead:", curr_f_dead,"\nc_m_dead:", curr_m_dead,"\nc_pop_alive:", curr_pop_alive,"\nc_pop_dead:", curr_pop_dead)        
# print()
# for x in range (0, 100000):
#     draw_sexy()
# print("curr_pop:",curr_pop,"\nc_f_dead:", curr_f_dead,"\nc_m_dead:", curr_m_dead,"\nc_pop_alive:", curr_pop_alive,"\nc_pop_dead:", curr_pop_dead)   
    
```

```python
#original test statistic
m_dead/f_dead
```

```python
final_means = []
for i in range(0,100):
    reset_sex()

    sample_size = int((genderdf['population'].sum())/1)
    observed_stat = m_dead/f_dead

    for j in range(0, 1000000):
        draw_sexy()

    if (len(picked_f) == 0):
        ratio = len(picked_m)/1
    else:
        ratio = len(picked_m)/len(picked_f)

    final_means.append(ratio)


#np.mean(final_means)
f = np.asarray(final_means)
tset, pval = ttest_1samp(f, observed_stat)
print(np.mean(final_means),m_dead/f_dead)
print('p-values',pval)
if pval < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
#final_means.mean()
```

#### Conclusion

The ttest gives us the p-value on the basis that our sample mean will not be much different from our population mean, that is, how authentic is our observed test statistic. Getting a value more than 0.05 means that we are accepting the null hypothesis which basically points to the fact that there is no significant difference between our sample mean our observed mean. 

Since the observed mean itself shows that males have a greater suicide rate than females, this proves the fact that males indeed have a greater suicide rate than females

## Hypothesis testing - Generation

```python
############################################ HYPOTHESIS TESTING FOR GENERATION ############################################
generation_df = pd.DataFrame(sdf, columns = ['year','country','generation', 'suicides_no', 'population'])
generation_df = generation_df[generation_df['year'] == 2014]
bool_arr = ((generation_df['country'] == 'United States'))
generation_df = generation_df[bool_arr]
generation_df = generation_df.groupby('generation').sum()

generation_df
```

```python
gen_pop_means = []

tot_pop = generation_df['population'].sum()
tot_dead = generation_df['suicides_no'].sum()

tot_boom = generation_df['suicides_no'][0]
gen_pop_means.append(tot_boom/(tot_pop + tot_dead))
#The below change was made only for Japan
tot_gi = 0 #generation_df['suicides_no'][1]
#gen_pop_means.append(tot_gi/(tot_pop + tot_dead))
tot_genX = generation_df['suicides_no'][1]
gen_pop_means.append(tot_genX/(tot_pop + tot_dead))
tot_genZ = generation_df['suicides_no'][2]
gen_pop_means.append(tot_genZ/(tot_pop + tot_dead))
tot_mill = generation_df['suicides_no'][3]
gen_pop_means.append(tot_mill/(tot_pop + tot_dead))
tot_silent = generation_df['suicides_no'][4]
gen_pop_means.append(tot_silent/(tot_pop + tot_dead))

curr_pop = 0;
curr_pop_alive = 0;
curr_pop_dead = 0;
curr_boom_dead = 0;
curr_gi_dead = 0;
curr_genX_dead = 0;
curr_genZ_dead = 0;
curr_mill_dead = 0;
curr_silent_dead = 0;

picked_boom = []
picked_gi = []
picked_genX = []
picked_genZ = []
picked_mill = []
picked_silent = []
picked_alive = []
```

```python
########################################### RESETTING #########################################
def reset_gen():
    
    global curr_pop_alive
    global curr_pop_dead
    global curr_boom_dead
    global curr_gi_dead
    global curr_genX_dead
    global curr_genZ_dead
    global curr_mill_dead
    global curr_silent_dead
    global curr_pop
    
    global tot_pop
    global tot_dead
    
    global tot_boom
    global tot_gi
    global tot_genX
    global tot_genZ
    global tot_mill
    global tot_silent
    
    global picked_boom
    global picked_gi
    global picked_genX
    global picked_genZ
    global picked_mill
    global picked_silent
    global picked_alive 
    
    curr_pop = tot_pop
    curr_pop_alive = tot_pop - tot_dead
    curr_pop_dead = tot_dead 
    curr_boom_dead = tot_boom;
    curr_gi_dead = tot_gi;
    curr_genX_dead = tot_genX;
    curr_genZ_dead = tot_genZ;
    curr_mill_dead = tot_mill;
    curr_silent_dead = tot_silent;
    
    picked_boom = []
    picked_gi = []
    picked_genX = []
    picked_genZ = []
    picked_mill = []
    picked_silent = []
    picked_alive = []

#################################################################################################################

def draw_gen():
    global curr_pop_alive
    global curr_pop_dead
    global curr_boom_dead
    global curr_gi_dead
    global curr_genX_dead
    global curr_genZ_dead
    global curr_mill_dead
    global curr_silent_dead
    global curr_pop
    
    global picked_boom
    global picked_gi
    global picked_genX
    global picked_genZ
    global picked_mill
    global picked_silent
    global picked_alive
    picked = 0
    
    #first of all pick a random number between 0 to 100.
    rando = random()
    rando = 0 + (rando * (100 - 0))
    # Now we check if this individual is going to commit suicide or nah
    # Chance for committing suicide is  = curr_people who died/ curr_popsize
    suicide_thresh = (curr_pop_dead/curr_pop) *100
    #print("thresh_s :",suicide_thresh, "rando:", rando)
    if rando <= suicide_thresh: 
        #we picked a dead person Prolly (boomer)
        picked =1
        #we need to determine who this is a dead 
        boom_thresh = (curr_boom_dead/ (curr_pop_dead)) * 100
        gi_thresh = boom_thresh + ((curr_gi_dead/ (curr_pop_dead)) * 100)
        genX_thresh = gi_thresh + ((curr_genX_dead/ (curr_pop_dead)) * 100)
        genZ_thresh = genX_thresh + ((curr_genZ_dead/ (curr_pop_dead)) * 100)
        mill_thresh = genZ_thresh + (curr_mill_dead/ (curr_pop_dead)) * 100
        
        rando_2 = random()
        rando_2 = 0 + (rando_2 * (100 - 0))
        
        if rando_2 <= boom_thresh:
            #this is a boomer, decrement boom dead pop and total pop
            curr_boom_dead = curr_boom_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_boom.append(picked)
        elif rando_2 <= gi_thresh:
            #this is a gi, decrement gi dead pop and total pop
            curr_gi_dead = curr_gi_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_gi.append(picked)
        elif rando_2 <= genX_thresh:
            #this is a genX, decrement genX dead pop and total pop
            curr_genX_dead = curr_genX_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_genX.append(picked)
        elif rando_2 <= genZ_thresh:
            #this is a genZ, decrement genZ dead pop and total pop
            curr_genZ_dead = curr_genZ_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_genZ.append(picked)
        elif rando_2 <= mill_thresh:
            #this is a mill, decrement mill dead pop and total pop
            curr_mill_dead = curr_mill_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_mill.append(picked)
        else:
            #this is a silent, decrement silent dead pop and total pop
            curr_silent_dead = curr_silent_dead -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_silent.append(picked)
    else:
        # we picked an alive person wuhuuuuu no one dies
        # update alive counter and total pop counter
        curr_pop_alive = curr_pop_alive-1
        curr_pop = curr_pop - 1
        picked_alive.append(picked)
        
##################################### testing (can ignore) ##############################################
# reset_gen()
# print("curr_pop:",curr_pop,"\nc_boom_dead:", curr_boom_dead,"\nc_curr_gi_dead:", curr_gi_dead, "\nc_genX_dead:", curr_genX_dead, "\nc_genZ_dead:", curr_genZ_dead,"\nc_mill_dead:", curr_mill_dead, "\nc_silent_dead:", curr_silent_dead,"\nc_pop_alive:", curr_pop_alive,"\nc_pop_dead:", curr_pop_dead)        
# print()
# for x in range (0, 100000):
#     draw_gen()
# print("curr_pop:",curr_pop,"\nc_boom_dead:", curr_boom_dead,"\nc_curr_gi_dead:", curr_gi_dead, "\nc_genX_dead:", curr_genX_dead, "\nc_genZ_dead:", curr_genZ_dead,"\nc_mill_dead:", curr_mill_dead, "\nc_silent_dead:", curr_silent_dead,"\nc_pop_alive:", curr_pop_alive,"\nc_pop_dead:", curr_pop_dead)
```

```python
counter = 0
picked_gen = 0

for k in gen_pop_means:
    final_means = []
    
    for i in range(0,10):
        reset_gen()

        sample_size = int((generation_df['population'].sum())/40000)
        observed_stat = k

        #for j in range(0, sample_size):
        for j in range(0, 1000000):
            draw_gen()
            
        #will have a populated picked alive, picked gen[i] and the corresponding current populations
        if counter == 0:
            picked_gen = len(picked_boom)
        #elif counter == 1:
        #    picked_gen = len(picked_gi)
        elif counter == 1:
            picked_gen = len(picked_genX)
        elif counter == 2:
            picked_gen = len(picked_genZ)
        elif counter == 3:
            picked_gen = len(picked_mill)
        else:
            picked_gen = len(picked_silent)
            
        #ratio = picked_gen/(len(picked_boom) + len(picked_gi) + len(picked_genX)+len(picked_genZ)+len(picked_mill)+len(picked_silent))
        ratio = picked_gen/1000000
        
        final_means.append(ratio)
    
    counter = counter+1

    #np.mean(final_means)
    f = np.asarray(final_means)
    tset, pval = ttest_1samp(f, observed_stat)
    print(np.mean(final_means),k)
    print('p-values',pval)
    if pval < 0.05:    # alpha value is 0.05 or 5%
        print(" we are rejecting null hypothesis")
    else:
        print("we are accepting null hypothesis")
```

#### Conclusion

```python
def reset_age():
    
    global curr_pop_alive
    global curr_pop_dead
    global curr_dead_15 
    global curr_dead_25 
    global curr_dead_35
    global curr_dead_5 
    global curr_dead_55 
    global curr_dead_75 
    global curr_pop
    global tot_pop
    global tot_dead
    global dead_15 
    global dead_25 
    global dead_35
    global dead_5 
    global dead_55 
    global dead_75 
    
    global picked_dead_15
    global picked_dead_25 
    global picked_dead_35 
    global picked_dead_5 
    global picked_dead_55 
    global picked_dead_75 
    global picked_alive
    
    curr_pop = tot_pop
    curr_pop_alive = tot_pop - tot_dead
    curr_pop_dead = tot_dead
    curr_dead_15 = dead_15
    curr_dead_25 = dead_25
    curr_dead_35 = dead_35
    curr_dead_5 = dead_5
    curr_dead_55 = dead_55
    curr_dead_75 = dead_75
    
    picked_dead_15 =[]
    picked_dead_25 = []
    picked_dead_35 = []
    picked_dead_5 = []
    picked_dead_55 = []
    picked_dead_75 = []
    picked_alive = []
    

########################################################################################

def draw_age():
    
    global curr_pop_alive
    global curr_pop_dead
    global curr_dead_15 
    global curr_dead_25 
    global curr_dead_35
    global curr_dead_5 
    global curr_dead_55 
    global curr_dead_75 
    global curr_pop

    global picked_dead_15
    global picked_dead_25 
    global picked_dead_35 
    global picked_dead_5 
    global picked_dead_55 
    global picked_dead_75
    global picked_alive
    
    picked = 0
    
    #first of all pick a random number between 0 to 100.
    randy = random()
    randy = 0 + (randy * (100 - 0))
    # Now we check if this individual is going to commit suicide or nah
    # Chance for committing suicide is  = curr_people who died/ curr_popsize
    suicide_thresh = (curr_pop_dead/curr_pop) *100
    #print("thresh_s :",suicide_thresh, "randy:", randy)
    if randy <= suicide_thresh: 
        #we picked a dead person
        picked =1
        # now we need to check what age group this person is belonging to
        randy_2 = random()
        randy_2 = 0 + (randy_2 * (100 - 0))
        
        # calculate thresh for each age
        thresh_15 = (curr_dead_15/ curr_pop_dead) * 100
        thresh_25 = (curr_dead_25/curr_pop_dead) * 100 + thresh_15
        thresh_35 = (curr_dead_35/ curr_pop_dead) * 100 + thresh_25
        thresh_5 = (curr_dead_5/ curr_pop_dead) * 100 + thresh_35
        thresh_55 = (curr_dead_55/curr_pop_dead) * 100 + thresh_5
        thresh_75 = (curr_dead_75/ curr_pop_dead) * 100 + thresh_55
        
        if randy_2 <= thresh_15:
            # select 15
            curr_dead_15 = curr_dead_15 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_15.append(picked)
            
        elif randy_2 <= thresh_25:
            #select 25
            curr_dead_25 = curr_dead_25 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_25.append(picked)
            
        elif randy_2 <= thresh_35:
            #select 35
            curr_dead_35 = curr_dead_35 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_35.append(picked)
            
        elif randy_2 <= thresh_5:
            #select 5
            curr_dead_5 = curr_dead_5 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_5.append(picked)
            
        elif randy_2 <= thresh_55:
            #select 55
            curr_dead_55 = curr_dead_55 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_55.append(picked)
        else:
            #select 75
            curr_dead_75 = curr_dead_75 -1
            curr_pop = curr_pop -1
            curr_pop_dead = curr_pop_dead -1
            picked_dead_75.append(picked)
    
    else:
        #picked alive
        curr_pop = curr_pop -1
        curr_pop_alive = curr_pop_alive -1
        picked_alive.append(picked)
        

##################################### testing (can ignore) ##############################################
# reset_age()
# print(curr_pop, curr_pop_alive, curr_pop_dead, curr_dead_15, curr_dead_25, curr_dead_35, curr_dead_5, curr_dead_55, curr_dead_75)
# for x in range (0, 1000000):
#     draw_age()
# #draw_age()
# print(curr_pop, curr_pop_alive, curr_pop_dead, curr_dead_15, curr_dead_25, curr_dead_35, curr_dead_5, curr_dead_55, curr_dead_75)
```
